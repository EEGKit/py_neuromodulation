### This script allows to first record user EEG and EMG data with OPenBCI headphones than train a classifier to detect different tasks

```python
%gui qt
from py_neuromodulation import (nm_settings, nm_define_nmchannels, nm_stream_offline, nm_train)
# from examples import (lange_n8_new)
from mne_lsl.lsl import resolve_streams
import pandas as pd
```

### Let's see if there are any streams on the network

```python
possible_streams = resolve_streams()
possible_streams
```

### Looks like we found the ExG-Headphone stream
### Let's find out more and setup a nm_stream

```python
exg_stream = possible_streams[0]
print(f'channel names: {exg_stream.get_channel_names()}')
print(exg_stream.get_channel_info)
```

```python
settings = nm_settings.get_default_settings()
settings["features"]["welch"] = False
settings["features"]["fft"] = True
settings["features"]["bursts"] = False
settings["features"]["sharpwave_analysis"] = False
settings["features"]["coherence"] = False


# create an array ch_names with the channel names "ch0", "ch1", "ch2", ...
ch_names = []
ch_types = []
for i in range(exg_stream.n_channels):
    ch_names.append(f'ch{i}')
    ch_types.append(exg_stream.stype)

nm_channels = nm_define_nmchannels.set_channels(
    ch_names = ch_names,
    ch_types= ch_types,
    reference = "default",
    new_names = "default",
    used_types= "eeg"
)

nm_channels
```

```python
stream = nm_stream_offline.Stream(
    sfreq=exg_stream.sfreq, nm_channels=nm_channels, settings=settings, verbose = True, line_noise = 50)
```
